```python
# --------- Global Variables -----------
import random
# who the winner is
winner = None
valid=[]
board ={
    (0,0):1,
    (0,1):2,
    (0,2):3,
    (1,0):4,
    (1,1):5,
    (1,2):6,
    (2,0):7,
    (2,1):8,
    (2,2):9
}
def display_board(board):
    '''This function print the board to the console'''
    print(' +-------+-------+-------+')
    print(' |  ',board[0,0],'\t','|  ',board[0,1],'\t','|  ',board[0,2],'\t','|',end='')
    print('\n','+-------+-------+-------+')
    print(' |  ',board[1,0],'\t','|  ',board[1,1],'\t','|  ',board[1,2],'\t','|',end='')
    print('\n','+-------+-------+-------+')
    print(' |  ',board[2,0],'\t','|  ',board[2,1],'\t','|  ',board[2,2],'\t','|',end='')
    print('\n','+-------+-------+-------+')
    print('\n')
def MakeListOfFreeFields(board):
    '''The function browses the board and builds a list of all the free squares'''
    valid=[]
    if board[(0,0)]!='X'and board[(0,0)]!='O':
        valid.append('1')
    if board[(0,1)]!='X'and board[(0,1)]!='O':
        valid.append('2')
    if board[(0,2)]!='X'and board[(0,2)]!='O':
        valid.append('3')
    if board[(1,0)]!='X'and board[(1,0)]!='O':
        valid.append('4')
    if board[(1,1)]!='X'and board[(1,1)]!='O':
        valid.append('5')
    if board[(1,2)]!='X'and board[(1,2)]!='O':
        valid.append('6')
    if board[(2,0)]!='X'and board[(2,0)]!='O':
        valid.append('7')
    if board[(2,1)]!='X'and board[(2,1)]!='O':
        valid.append('8')
    if board[(2,2)]!='X'and board[(2,2)]!='O':
        valid.append('9')
    print('The valid locations is:',str(valid))  
    return(valid)
def DrawComputerMove(board):
    '''The function draws the computer's move and updates the board'''
    mov=random.randint(1,9)
    print('Computer choosed:',mov,'\n')
    
    if len(valid)!=0:
        if mov==1 and (board[(0,0)] !="X" and board[(0,0)] != "O"):
            board[(0,0)]="O"
           
        elif mov==2 and (board[(0,1)] !="X" and board[(0,1)] != "O"):
            board[(0,1)]="O"
        elif mov==3 and (board[(0,2)] !="X" and board[(0,2)] != "O"):
            board[(0,2)]="O"
        elif mov==4 and (board[(1,0)] !="X" and board[(1,0)] != "O"):
            board[(1,0)]="O"
        elif mov==5 and (board[(1,1)] !="X" and board[(1,1)] != "O"):
            board[(1,1)]="O"
        elif mov==6 and (board[(1,2)] !="X" and board[(1,2)] != "O"):
            board[(1,2)]="O"
        elif mov==7 and (board[(2,0)] !="X" and board[(2,0)] != "O"):
            board[(2,0)]="O"
        elif mov==8 and (board[(2,1)] !="X" and board[(2,1)] != "O"):
            board[(2,1)]="O"
        elif mov==9 and (board[(2,2)] !="X" and board[(2,2)] != "O"):
            board[(2,2)]="O"
        return(board)
def EnterUserMove(board):
    '''Asks the user about their move, checks the input, and updates the board according to the user's decision'''
    mov=int(input('Your turn '))
    print('You choosed:',mov,'\n')
    if len(valid)!=0:
        if mov==1 and (board[(0,0)] !="X" and board[(0,0)] != "O"):
            board[(0,0)] = "X"
            
        elif mov==2 and ( board[(0,1)] !="X" and board[(0,1)] != "O"):
            board[(0,1)]="X"
            
        elif mov==3 and (board[(0,2)] !="X" and board[(0,2)] != "O"):
            board[(0,2)]="X"
            
        elif mov==4 and (board[(1,0)] !="X" and board[(1,0)] != "O"):
            board[(1,0)]="X"
            
        elif mov==5 and (board[(1,1)] !="X" and board[(1,1)] != "O"):
            board[(1,1)]="X"
            
        elif mov==6 and (board[(1,2)] !="X" and board[(1,2)] != "O"):
            board[(1,2)]="X"
            
        elif mov==7 and (board[(2,0)] !="X" and board[(2,0)] != "O"):
            board[(2,0)]="X"
            
        elif mov==8 and (board[(2,1)] !="X" and board[(2,1)] != "O"):
            board[(2,1)]="X"
            
        elif mov==9 and (board[(2,2)] !="X" and board[(2,2)] != "O"):
            board[(2,2)]="X"    
    return(board )
def VictoryFor(board):
    '''the function analyzes the board status in order to check if the player using 'O's or 'X's has won the game'''
    #global winner
    
    #row_winner
    if board[(0,1)]==board[(0,2)]==board[(0,0)]=='X':
        winner='x'
        return(winner)
    elif board[(0,1)]==board[(0,2)]==board[(0,0)]=='O':
        winner='O'
        return(winner)
    elif board[(1,0)]==board[(1,2)]==board[(1,1)]=='X':
        winner='X'
        return(winner)
    elif board[(1,0)]==board[(1,2)]==board[(1,1)]=='O':
        winner='O'
        return(winner)
    elif board[(2,0)]==board[(2,2)]==board[(2,1)]=='X':
        winner='X'
        return(winner)
    elif board[(2,0)]==board[(2,2)]==board[(2,1)]=='O':
        winner='O'
        return(winner)
    #column_winner
    elif board[(1,0)]==board[(2,0)]==board[(0,0)]=='X':
        winner='X'
        return(winner)
    elif board[(0,0)]==board[(1,0)]==board[(2,0)]=='O':
        winner='O'
        return(winner)
    elif board[(1,1)]==board[(0,1)]==board[(2,1)]=='X':
        winner='X'
        return(winner)
    elif board[(1,1)]==board[(0,1)]==board[(2,1)]=='O':
        winner='O'
        return(winner)
    elif board[(0,2)]==board[(1,2)]==board[(2,2)]=='X':
        winner='X'
        return(winner)
    elif board[(0,2)]==board[(1,2)]==board[(1,2)]=='O':
        winner='O'
        return(winner)
    #diagonals_winner
    elif board[(0,0)]==board[(2,2)]==board[(1,1)]=='X':
        winner='X'
        return(winner)
    elif board[(0,0)]==board[(2,2)]==board[(1,1)]=='O':
        winner='O'
        return(winner)
    elif board[(2,0)]==board[(0,2)]==board[(1,1)]=='X':
        winner='X'
        return(winner)
    elif board[(2,0)]==board[(0,2)]==board[(1,1)]=='O':
        winner='O'
        return(winner)
    
def Tie(valid):
    if len(valid)==0 :
        print('Game finished with Tie')
        return False
    else:
        return True
def Who_win(winner):
    if winner=='X':
        print('you Won')
        return False
    elif winner=='O':
        print('You Lost')
        return False
    else:
        return True
while True:
    display_board(board)
    DrawComputerMove(board)  
    MakeListOfFreeFields(board)
    VictoryFor(board)
    display_board(board)
    Tie(valid)
    Who_win(winner)
    display_board(board)
    EnterUserMove(board)
    MakeListOfFreeFields(board)
    VictoryFor(board)
    display_board(board)
    Tie(valid)
    Who_win(winner)
            
        
    
```
